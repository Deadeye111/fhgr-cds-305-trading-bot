```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

# Dictionary to store the sector information
stock_sector_mapping = {}

# Fetch sector information for each stock
for ticker in stocks:
    stock = yf.Ticker(ticker)
    info = stock.info
    sector = info.get('sector', 'Unknown')
    stock_sector_mapping[ticker] = sector

# Count the number of stocks in each sector
sector_counts = Counter(stock_sector_mapping[ticker] for ticker in stocks)

# Convert the counter to a pandas DataFrame
sector_df = pd.DataFrame.from_dict(sector_counts, orient='index', columns=['Count'])

# Plot the bar chart
plt.figure(figsize=(10, 6))
sector_df['Count'].plot(kind='bar', color='skyblue')
plt.title('Number of Stocks in Each Sector')
plt.xlabel('Sector')
plt.ylabel('Number of Stocks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot the pie chart
plt.figure(figsize=(8, 8))
sector_df['Count'].plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(sector_df))))
plt.title('Distribution of Stocks by Sector')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
