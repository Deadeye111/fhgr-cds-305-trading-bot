# Crypto Analysis


```python
# Download bitcoin data
ada = yf.download('ADA-USD', start= '2023-01-01', end='2024-3-31')
atom = yf.download('ATOM-USD', start= '2023-01-01', end='2024-3-31')
avax = yf.download('AVAX-USD', start= '2023-01-01', end='2024-3-31')
bch = yf.download('BCH-USD', start= '2023-01-01', end='2024-3-31')
bnb = yf.download('BNB-USD', start= '2023-01-01', end='2024-3-31')
btc = yf.download('BTC-USD', start= '2023-01-01', end='2024-3-31')
doge = yf.download('DOGE-USD', start= '2023-01-01', end='2024-3-31')
dot = yf.download('DOT-USD', start= '2023-01-01', end='2024-3-31')
eth = yf.download('ETH-USD', start= '2023-01-01', end='2024-3-31')
hbar = yf.download('HBAR-USD', start= '2023-01-01', end='2024-3-31')
link = yf.download('LINK-USD', start= '2023-01-01', end='2024-3-31')
shib = yf.download('SHIB-USD', start= '2023-01-01', end='2024-3-31')
sol = yf.download('SOL-USD', start= '2023-01-01', end='2024-3-31')
trx = yf.download('TRX-USD', start= '2023-01-01', end='2024-3-31')
xrp = yf.download('XRP-USD', start= '2023-01-01', end='2024-3-31')
```

```python
# merge all dataframes into one and only use the 'Close' column
df = pd.concat([ada['Close'], atom['Close'], avax['Close'], bch['Close'], bnb['Close'], btc['Close'], doge['Close'], dot['Close'], eth['Close'], hbar['Close'], link['Close'], shib['Close'], sol['Close'], trx['Close'], xrp['Close']], axis=1)
df.columns = ['ADA', 'ATOM', 'AVAX', 'BCH', 'BNB', 'BTC', 'DOGE', 'DOT', 'ETH', 'HBAR', 'LINK', 'SHIB', 'SOL', 'TRX', 'XRP']

# data as index
df.index = pd.to_datetime(df.index)
```

```python
df
```

```python
# normalize the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

# add dates from the original dataframe as index
df_scaled.index = df.index

```

```python
df_scaled
```

```python
# plot the data
df_scaled.plot(figsize=(10, 6))
plt.title('Kryptopreise im Testzeitraum (skaliert)')
#legend position
plt.legend(loc='upper left')
```
